```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/performance_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#plt.figure(figsize=(10, 6))

plt.boxplot([df['ER'], df['CTR']], labels=['ER', 'CTR'])
plt.title('Boxplot of ER and CTR')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show();
```

```python
sns.histplot(df['ER'], bins=20, kde=True, color='red', label='ER')
sns.histplot(df['CTR'], bins=20, kde=True, color='blue', label='CTR')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of ER and CTR')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='CTR', y='ER', data=df)
plt.xlabel('CTR')
plt.ylabel('ER')
plt.title('The relationship between CTR and ER')
plt.show();
```
